## <span style = "color:orange"> Visualization </span>

### Plot a scatterplot of Avg_credit_Limit vs Total_Credit_Cards with respect to Total_calls_made

```python
sns.scatterplot(x='Avg_Credit_Limit',y='Total_Credit_Cards',hue='Total_calls_made',data=df)
```

### Plot correlation matix on Heatmap

```python
sns.heatmap(df.corr(),annot=True)
```

### Plot a pairplot

```python
sns.pairplot(df)
```

### Create K Means model (Let k = 4)

```python
from sklearn.cluster import KMeans
```

```python
model=KMeans(n_clusters=4)
```

### Fit predict for our Dataset

```python
cluster=model.fit_predict(df)
cluster
```

### Assign the predicted clusters to a new column in our dataset

```python
df['Clusters']=model.fit_predict(df)
```

```python
df.head()
```

## Visualize the clusters

```python
# Use x = 'Avg_Credit_Limit' ang y =  'Total_Credit_Cards'
```

```python
df['Clusters'].unique()
```

```python
df0 = df[df['Clusters'] == 0]
df1 = df[df['Clusters'] == 1]
df2 = df[df['Clusters'] == 2]
df3 = df[df['Clusters'] == 3]
```

```python
sns.scatterplot(x = 'Avg_Credit_Limit',y =  'Total_Credit_Cards',data=df0,color = 'green')
sns.scatterplot(x = 'Avg_Credit_Limit',y =  'Total_Credit_Cards',data=df1,color = 'blue')
sns.scatterplot(x = 'Avg_Credit_Limit',y =  'Total_Credit_Cards',data=df2,color = 'yellow')
sns.scatterplot(x = 'Avg_Credit_Limit',y =  'Total_Credit_Cards',data=df3,color = 'red')
plt.show()
```
